```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def preprocess(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation (optional: you might want to keep certain punctuation)
    text = ''.join([char for char in text if char.isalnum() or char.isspace()])
    return text

def remove_duplicates(questions, threshold=0.9):
    # Vectorize the questions
    vectorizer = TfidfVectorizer(stop_words='english', tokenizer=None, preprocessor=preprocess)
    tfidf_matrix = vectorizer.fit_transform(questions)
    
    # Calculate cosine similarity matrix
    cos_sim_matrix = cosine_similarity(tfidf_matrix)
    
    # Find indices of duplicates based on the threshold
    duplicates = set()
    for i in range(cos_sim_matrix.shape[0]):
        for j in range(i+1, cos_sim_matrix.shape[1]):
            if cos_sim_matrix[i, j] > threshold:
                duplicates.add(j)
    
    # Filter out duplicates
    filtered_indices = [i for i in range(len(questions)) if i not in duplicates]
    return questions.iloc[filtered_indices], duplicates

def main():
    # Load data from CSV
    file_path = 'path_to_your_file.csv'  # Update this path
    df = pd.read_csv(file_path)
    
    # Assuming the questions are under the column name 'Question'
    filtered_questions = remove_duplicates(df['Question'])
    
    # Save the filtered data
    df.iloc[filtered_questions.index].to_csv('filtered_questions.csv', index=False)
    print("Filtered questions saved to 'filtered_questions.csv'.")

# if __name__ == "__main__":
#     main()
```

```python
vectorizer = TfidfVectorizer(stop_words='english', tokenizer=None, preprocessor=preprocess)
question_data = data[data['question_type']=='saq'].reset_index(drop=True)
questions = (question_data['question']).apply(preprocess)
tfidf_matrix = vectorizer.fit_transform(questions)
```

```python
cos_sim_matrix = cosine_similarity(tfidf_matrix)
cos_sim_matrix
```

```python
pd.DataFrame(cos_sim_matrix).iloc[:10, :10]
```

```python
def identify_and_sort_duplicates(cos_sim_matrix, threshold=0.75):
    duplicates = []
    for i in range(cos_sim_matrix.shape[0]):
        for j in range(i+1, cos_sim_matrix.shape[1]):
            if cos_sim_matrix[i, j] > threshold:
                duplicates.append((i, j, cos_sim_matrix[i, j]))

    # Sort duplicates by decreasing similarity score
    duplicates_sorted = sorted(duplicates, key=lambda x: x[2], reverse=True)
    return duplicates_sorted

duplicates_sorted = identify_and_sort_duplicates(cos_sim_matrix)
```
